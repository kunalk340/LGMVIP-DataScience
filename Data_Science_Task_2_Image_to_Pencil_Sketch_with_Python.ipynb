# Task-2 - Image to Pencil Sketch with Python

### Level -  Beginner Level 

```python
# Lets import the required library.
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

### CREATING IMAGES


```python
# Reading the original image
img = cv2.imread('/content/sample_data/tiger.jpg')
cv2_imshow(img)
```

```python
#Convert the RGB image to grayscal image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
#Invert the Grayscale Image
invert_img = 225-gray_img
cv2_imshow(invert_img)
```

```python
#Smoothen the innverted image using gaussian blur technique.
img_smoothen = cv2.GaussianBlur(invert_img, (21, 21), sigmaX = 0, sigmaY = 0)
cv2_imshow(img_smoothen)
```

```python
# Now create a pencil Sketch image
inverted_blurr = 255-img_smoothen
pencil_sketch = cv2.divide(gray_img, invert_img, inverted_blurr, scale = 256.0)
cv2_imshow(pencil_sketch)
```

```python
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```

### Conclusion

#### So in this way using opencv we can get pencil sketch of an image
